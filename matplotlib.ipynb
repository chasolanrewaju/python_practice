```python
import matplotlib.pyplot as plt

x = [2,4,6,8,10]
y = [10,20,30,40,50]

# diagram size
plt.figure (figsize=(4,4), dpi=150)

plt.plot (x,y, label = 'Buhari', color = 'green',linewidth = 2, linestyle = '--', marker = '.',markersize = 15  )

# Title naming and editing
plt.title ('My first Graph',fontdict={'fontname' : 'Times new Roman', 'fontsize': 20} )
plt.xlabel ('x axis')
plt.ylabel ('y axis')

plt.legend()

#saving generated figure
plt.savefig ('graphimage.png',dpi= 150 )
plt.show()
```
